<h2 style = "color:orange; font-size: 25px;">Basic Equations</h2>

```python
# library to display latex 
from IPython.display import Math, display

# declare an equation
eq = (x^2 - 4 == 0)

# solve(eq,x) solves for x in eq
ans = solve(eq,x)

# latex(eq) returns latex of eq
ans = latex(ans)

# Math() displays latex in pretty format
Math(ans)
```

<h2 style = "color:orange; font-size: 25px;">Writing Algebra</h2>

```python
# no matter how you structure the equation, it will be displayed from highest to lowest in term degrees
# so even though it is written as 2x^2 -6 + 4x, still shows as 2x^2 + 4x - 6
eq = 2*x^2 -6 + 4*x

# an equation can be factored:
factored_form = latex(eq.factor())
display(Math(latex(eq)))
print('->')
display(Math(factored_form))

# '\\text{}' is used to write text in latex
display(Math('\\text{Here is equation: }' + factored_form))

# textbf bolds text
display(Math('\\textbf{I am a bold text: }' + factored_form))

# textit italicises text
display(Math('\\textit{I am an italicised text: }' + factored_form))
```
