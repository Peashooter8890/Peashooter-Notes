<h2 style = "color:orange; font-size: 25px;">Basic 2D Plot</h2>

```octave
# Plot in the form start:steps:end
# Without ; it will show every single data point in numerical form
x = 1:1:20;

# For exponents, we have to use .^ instead of ^ because syntax (langauge is oriented around matrices)
y = x.^2;

# Plotting is (x,y,color). Color is blue by default
plot(x,y,"r") 

# We can set title and labels here. 
title("basic plot")
xlabel("x-axis")
ylabel("y-axis")
```

<h2 style = "color:orange; font-size: 25px;">Multiple 2D Plots</h2>

<p>Simply attempting to display multiple graphs one-by-one will result in a failure - it will only show the final one.</p>

```octave
x = 1:1:20;
x1 = 1:1:20;
y = x;
y2 = x.^2;

# Here we attempt to plot two different graphs of different colors, but only the last one (green) shows
plot(x,y,"r") 
plot(x1,y2,"g") 
```

<p>This is because by default, new graphs will replace the existing one. Addressing this issue is simple - you have to designate a period in which this property is overwritten. <b>hold on</b> and <b>hold off</b> can do that. </p>

```octave
x = 1:1:20;
x1 = 1:1:20;
x2 = 1:1:20;
y = x;
y1 = x.^2;
y2 = x + 10;

plot(x,y,"r") 
# Tell computer: don't replace any graphs that come from now or before!
hold on;
plot(x1,y1,"g") 
plot(x2,y2,"b") 
# Tell computer: you can replace graphs from now on!
hold off;
```

<h2 style = "color:orange; font-size: 25px;">Manipulation of 2D Plots</h2>

```octave
 clf;
 r = 3;
 c = 3;
 fmt = {"horizontalalignment", "center", "verticalalignment", "middle"};
 for n = 1 : r*c
   subplot (r, c, n);
    xlabel (sprintf ("xlabel #%d", n));
    ylabel (sprintf ("ylabel #%d", n));
    axis ([0 1 0 1]);
 endfor
 subplot (r, c, 1:3);
  xlabel (sprintf ("xlabel #%d:%d", 1, 3));
  ylabel (sprintf ("ylabel #%d:%d", 1, 3));
 axis ([0 1 0 1]);
```
